```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
img=cv2.imread("colors.webp")

img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
cv2.waitKey(0)
```

```python
cv2.imwrite("gray_image.jpg", img_gray)
```
